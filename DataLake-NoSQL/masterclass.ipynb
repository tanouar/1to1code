## Sécurite des identifiants de connexion (credentials)

La manière la plus simple de stocker vos identifiants de connexion est de les stocker dans des variables d'environnement. Pour cela nous allons crée un fichier `.env` à la racine de notre projet, et y stocker nos identifiants de connexion. Par exemple pour AWS nous allons stocker notre `AWS_ACCESS_KEY_ID` et notre `AWS_SECRET_ACCESS_KEY`, pour MongoDB Atlas nous allons stocker notre `MONGODB_URI`, et pour Redis nous allons stocker notre `HOSTNAME`, `PORT`, `PASSWORD`

Voici un exemple de fichier `.env` :

```env
# Identifiants AWS
AWS_ACCESS_KEY_ID=your_access_key_id
AWS_SECRET_ACCESS_KEY=your_secret_access_key
AWS_DEFAULT_REGION =your_aws_region

# Identifiants MongoDB Atlas
MONGODB_USER=your_mongodb_user
MONGODB_PASSWORD=your_mongodb_password
MONGODB_DB=your_mongodb_db

# Identifiants Redis
REDIS_HOSTNAME=your_redis_hostname
REDIS_PORT=your_redis_port
REDIS_PASSWORD=your_redis_password
```

Deux possibilité, soit vous crée un fichier `.env` dans le même repertoire que votre script python ou notebook vous même ou alors vous pouver exécuter la commande suivante dans votre terminal pour créer le fichier `.env` et y ajouter les variables d'environnement :

```bash
echo "AWS_ACCESS_KEY_ID=your_access_key_id" >> .env
echo "AWS_SECRET_ACCESS_KEY=your_secret_access_key" >> .env
echo "AWS_DEFAULT_REGION=your_aws_region" >> .env
echo "MONGODB_USER=your_mongodb_user" >> .env
echo "MONGODB_PASSWORD=your_mongodb_password" >> .env
echo "MONGODB_DB=your_mongodb_db" >> .env
echo "REDIS_HOSTNAME=your_redis_hostname" >> .env
echo "REDIS_PORT=your_redis_port" >> .env
echo "REDIS_PASSWORD=your_redis_password" >> .env
```
